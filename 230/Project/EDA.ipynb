```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("SeoulBikeData.csv", encoding="unicode_escape")

# Convert column names for easier access
df.columns = df.columns.str.strip()

# Plot 1: Average hourly bike rentals
hourly_avg = df.groupby("Hour")["Rented Bike Count"].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(data=hourly_avg, x="Hour", y="Rented Bike Count", marker="o")
plt.title("Average Hourly Bike Rentals")
plt.xlabel("Hour of Day")
plt.ylabel("Average Rented Bike Count")
plt.grid(True)
plt.xticks(range(0, 24))
plt.tight_layout()
plt.show()

# Plot 2: Correlation heatmap of all numeric variables
# Select only numeric columns
numeric_cols = df.select_dtypes(include=["float64", "int64"])

# Compute correlation matrix
corr_matrix = numeric_cols.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.title("Correlation Heatmap of Numeric Variables")
plt.tight_layout()
plt.show()
```
